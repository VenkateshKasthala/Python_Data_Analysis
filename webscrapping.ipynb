```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url1 = 'https://en.wikipedia.org/wiki/Pawan_Kalyan_filmography'
url2 = 'https://en.wikipedia.org/wiki/Mahesh_Babu_filmography'

response1 = requests.get(url1)
response2 = requests.get(url2)


soup1 = BeautifulSoup(response1.content, 'html.parser')
soup2 = BeautifulSoup(response2.content, 'html.parser')

table1 = soup1.find_all('table')[1]

table2 = soup2.find_all('table')[2]

data1 = []
for row in table1.find_all('tr'):
    row_data = [cell.text.strip() for cell in row.find_all(['td', 'th'])]
    data1.append(row_data)
    
data2 = []
for row in table2.find_all('tr'):
    row_data = [cell.text.strip() for cell in row.find_all(['td', 'th'])]
    data2.append(row_data)
```

```python
df1 = pd.DataFrame(data1[1:], columns=data1[0]) 
df1['id']=1
df1['actor']='Pawan Kalyan'

df2 = pd.DataFrame(data2[1:], columns=data2[0])
df2['id']=2
df2['actor']='Mahesh Babu'
```

```python
df1 = df1.rename(columns={'Title' :'Film','Role' :'Role(s)'})
df1.loc[[3,9,12,17], 'Year'] = [1998,2004,2006,'']


df1
```
